# Image Scraping

```python
#importing libraries
from selenium import webdriver
import pandas as pd
import warnings   #importing warning modules for ignoring the warnings
warnings.filterwarnings('ignore')
import time
import requests
```

```python
#connecting to webdriver
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
#opening the url
driver.get('https://www.amazon.in/')
time.sleep(2)
driver.maximize_window()
```

```python
search=driver.find_element_by_id("twotabsearchtextbox")
time.sleep(3)
search.send_keys("sarees")
```

```python
button=driver.find_element_by_id("nav-search-submit-button")
time.sleep(1)
button.click()
```

```python
page_link="https://www.amazon.in/s?k=sarees&crid=7UTQGCRKWP2&qid=1648436029&sprefix=sarees%2Caps%2C503&ref=sr_pg_"
```

```python
img_saree=[]

#connecting to each pages
for next_page in range(1,15):
    next_page=page_link + str(next_page)
    #connecting to each page
    driver.get(next_page)
    
    for i in driver.find_elements_by_xpath("//img[@class='s-image']"):
        img_saree.append(i.get_attribute('src'))
```

```python
len(img_saree)
```

```python
driver.close()
```
